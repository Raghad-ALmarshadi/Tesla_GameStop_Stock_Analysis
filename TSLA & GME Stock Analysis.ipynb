```python
# 2.2 Revenue data using Web Scraping
import yfinance as yf
import pandas as pd

# Tesla Data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="5y")
tesla_data.reset_index(inplace=True)

# Create "Approximate Revenue" using Close Price multiplied by Volume
tesla_revenue = pd.DataFrame({
    "Date": tesla_data['Date'],
    "Revenue": tesla_data['Close'] * tesla_data['Volume']
})

tesla_revenue.tail()

```

```python
# 3.2 Revenue data using Web Scraping
import yfinance as yf
import pandas as pd

# GameStop Data
gamestop = yf.Ticker("GME")
gamestop_data = gamestop.history(period="5y")
gamestop_data.reset_index(inplace=True)

# Create "Approximate Revenue" using Close Price multiplied by Volume
tesla_revenue = pd.DataFrame({
    "Date": tesla_data['Date'],
    "Revenue": tesla_data['Close'] * tesla_data['Volume']
})

tesla_revenue.tail()

```

```python
# ---------------------------------------------
# 4️⃣ Merge data and create Dashboard
# ---------------------------------------------

# 4.1 merge Tesla
tesla_dashboard = tesla_data.merge(tesla_revenue, how="left", left_on="Date", right_on="Date")
# Convert Revenue to Number
tesla_dashboard['Revenue'] = pd.to_numeric(tesla_dashboard['Revenue'], errors='coerce')

fig_tesla = px.line(tesla_dashboard, x="Date", y=["Close", "Revenue"],
                    title="Tesla Stock Price vs Revenue",
                    labels={"value":"Amount (USD)", "variable":"Metric"})
fig_tesla.show()
```

```python
# 4.2 merge GameStop
gme_dashboard = gamestop_data.merge(gme_revenue, how="left", left_on="Date", right_on="Date")
gme_dashboard['Revenue'] = pd.to_numeric(gme_dashboard['Revenue'], errors='coerce')

fig_gme = px.line(gme_dashboard, x="Date", y=["Close", "Revenue"],
                  title="GameStop Stock Price vs Revenue",
                  labels={"value":"Amount (USD)", "variable":"Metric"})
fig_gme.show()
```
